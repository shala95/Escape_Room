```python
import os
import pandas as pd
import numpy as np
from scipy.io import loadmat

def load_patient_data_to_dataframe(ROOT_PATH):
    all_data = []

    for filename in os.listdir(ROOT_PATH):
        if filename.endswith('.mat') and not filename.endswith('_ann.mat') and not filename.endswith('_spk.mat'):
            patient_id = filename.split('.')[0]
            mat_file = os.path.join(ROOT_PATH, filename)
            ann_file = os.path.join(ROOT_PATH, patient_id + '_ann.mat')
            spk_file = os.path.join(ROOT_PATH, patient_id + '_spk.mat')

            try:
                # Load PPG signal
                ppg_data = loadmat(mat_file)['ppg'].flatten()

                # Load labels
                labels_data = loadmat(ann_file)['labels'].flatten() if os.path.exists(ann_file) else np.full(ppg_data.shape, None)

                # Load peak positions
                speaks_data = loadmat(spk_file)['speaks'].flatten() if os.path.exists(spk_file) else np.full(ppg_data.shape, None)

                for ppg, label, peak in zip(ppg_data, labels_data, speaks_data):
                    all_data.append({
                        'PatientID': patient_id,
                        'PPG': ppg,
                        'Label': label,
                        'Peak': peak
                    })

                print(f"Loaded data for patient {patient_id}")
            except Exception as e:
                print(f"Error loading data for patient {patient_id}: {e}")

    # Convert list of dictionaries to DataFrame
    df = pd.DataFrame(all_data)
    return df

# Example usage
# folder_path = '/content/drive/MyDrive/ai_project_data'
patients_df1 = load_patient_data_to_dataframe(ROOT_PATH)

```
